```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import joblib
```

```python
all_data = []

for year in years:
    try:
        df_com = pd.read_excel('/content/SupplyChainEmissionFactorsforUSIndustriesCommodities.xlsx', sheet_name=f'{year}_Detail_Commodity')
        df_ind = pd.read_excel('/content/SupplyChainEmissionFactorsforUSIndustriesCommodities.xlsx', sheet_name=f'{year}_Detail_Industry')

        df_com['Source'] = 'Commodity'
        df_ind['Source'] = 'Industry'
        df_com['Year'] = df_ind['Year'] = year

        df_com.columns = df_com.columns.str.strip()
        df_ind.columns = df_ind.columns.str.strip()

        df_com.rename(columns={
            'Commodity Code': 'Code',
            'Commodity Name': 'Name'
        }, inplace=True)

        df_ind.rename(columns={
            'Industry Code': 'Code',
            'Industry Name': 'Name'
        }, inplace=True)

        all_data.append(pd.concat([df_com, df_ind], ignore_index=True))

    except Exception as e:
        print(f"Error processing year {year}: {e}")
```

```python
df = pd.concat(all_data, ignore_index=True)
df.head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Substance', y='Supply Chain Emission Factors with Margins', data=df)
plt.title('Distribution of Supply Chain Emission Factors with Margins by Substance')
plt.xlabel('Substance')
plt.ylabel('Emission Factor with Margins')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Supply Chain Emission Factors with Margins'], kde=True)
plt.title('Distribution of Supply Chain Emission Factors with Margins')
plt.xlabel('Emission Factor with Margins')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Source', y='Supply Chain Emission Factors with Margins', data=df)
plt.title('Distribution of Supply Chain Emission Factors with Margins by Source')
plt.xlabel('Source')
plt.ylabel('Emission Factor with Margins')
plt.tight_layout()
plt.show()
```
